In this project, a XRAY transformer with the given specifications below will be designed:

- Single Phase, High Frequency High Voltage Transformer
- Primary Winding Voltage ± 417 V (peak to peak 834 V for pulsing)
- Secondary Winding Voltage ± 12.5 kV (peak to peak 25 kV for pulsing)
- Rated Power 30 kW (for maximum 100 millisecond)
- Switching Frequency Minimum 100 kHz
- Ambient Temperature 0-40 °C


```python
from IPython.display import Image
Image(filename='MagnetisationGraphics.png', width=550, height=550)
```

Now, let's define input parameters of the transformer:

```python
import math

V_in=417 #input (primary) voltage in Volts
V_out=12500 #output (secondary) voltage in Volts
P_out=30000 #rated output power in Watts
f=100000 #frequency in Hz
T_min=0 #min ambient temperature in Celcius
T_max=40 #max ambient temperature in Celcius
```

Constants:

```python
Ku=0.35 #window utilization factor
Kf=4 #square wave coefficient
```

For the first transformer design iteration, initalize the parameters as below:

```python
Bop=0.35 #operating flux denisty in Tesla
eff=0.90 #efficiency
J=3 # current density in A/mm2
```

Power handling capability of a core is related to its area product Ap. To find Ap:

```python
P_in=P_out/eff
P_total=P_in+P_out #total apparent power of the transformer

Ap=(P_total)/(Bop*f*Ku*Kf*J*1e6) #in
print (Ap*1e8) #in cm4
```

From the ferrite core catalog in http://www.ferroxcube.com/, E shape E80/38/20 core is selected in accordance with 43 cm4 area product. Core dimensions are given below:

```python
from IPython.display import Image
Image(filename='E80-38-20.png', width=550, height=550)
```

```python
Ae=392 #effectice core area in mm2 according to core datasheet
```

```python
Np=(V_in*1e6)/(Kf*f*Bop*Ae)#Primary number of turns using Faraday's Law
print (Np)
```

Number of turns at secondary side can be found as follows:

```python
Ns=(V_out*Np)/(V_in)
print (Ns)
```

We can approximate Np and Ns as below:

```python
Np=7
Ns=227
```

Now it is time to determine the cable sizes to use in primary and secondary windings. For this purpose, skin depth has to be calculated.

```python
resistivity=1.69*1e-8 #copper resistivity in ohm.m
rel_per=1*4*math.pi*1e-7 #relative permeability of ferrite core
Skin_depth=math.sqrt((resistivity)/(math.pi*f*rel_per))*1e2 #skin depth in cm 
print (round(Skin_depth,4)) #skin depth in cm
Cop_area=math.pi*Skin_depth*Skin_depth
print (Cop_area)
```

From the AWG list, AWG26 cable area is sufficient to cover our skin depth. Therefore AWG26 cable is chosen. Now, let's find the current to be carried in the primary winding. In single AWG26 cable, current conducting capacity is:

```python
I_cap=Cop_area*J*1e2
print(I_cap)
```

```python
I_p=P_in/V_in
print (I_p)
```

Number of stranded conductors to be needed in the primary side:

```python
num_str=I_p/I_cap
print (num_str)
```

Area of 199 stranded conductor:

```python
dia_awg26=0.405
Str_area=199*math.pi*dia_awg26*dia_awg26/4
print (Str_area)
```

```python
Str_dia=2*math.sqrt(Str_area/math.pi)
print (Str_dia) # in mm
```

Calculate number of stranded conductors in the secondary side LATER!

Let's calculate the stranded conductor resistance: 

```python
Str_res=133.9/num_str #in mohm per meter
print (Str_res)
```

In order to find the length of the cable to be wrap around primary side, first calculate the perimeter of the core:

```python
per_core=19.8*4 #perimeter in mm
print (per_core)
cable_length=per_core*Np
print (cable_length) #in mm
cable_resistance=cable_length*1e-3*Str_res
print (cable_resistance) # in mohm
```
