** Determination of main dimensions **

Althuogh it is hard to enqueue the steps of the motor design because of its being an iterative process, it is a good choice to start with determining the main machine dimensions according to standards and past experiences.

In the first step, we will set airgap diameter or and the axial length of the core. There is a parameter, called Cmech (machine constant), which indicates electrical and magnetic loading of the machine and can take values from the following interval:

```python
from IPython.display import Image 
Image(filename='Cmech.PNG')
```

In the above graph, Cmech values with respect to pole-power are given.

```python
pole_power = Pmech/(2*polepair)
print 'Pole Power = ',pole_power/1000.0,'kW'
```

From the graph, it is seen that we can start with Cmech = 250 kW.s/m^3.

```python
Cmech = 250000.0 # W.s/m^3
```

Cmech is related to machine dimensions with the expression: Cmech = Pmech / (D^2 x lp x nsynch)

Pmech is the mechanical power.

D is the stator inner diameter.

lp is equivalent axial length.

nsynch is the synchronous rotor speed in Hz.

```python
nsynch = f/polepair
```

To be able to find D and l value from the above equation, we need to know aspect ratio of the motor, which is X = l/D. Aspect ratio for a 6 pole motor is given by this general equation: 

X = (pi/2*polepair)*polepair^1/3

```python
X=(pi/(2*polepair))*(polepair**(1.0/3.0))
Dis = (Pmech/(Cmech*X*nsynch))**(1.0/3.0)
Dis = round(Dis*100.0)/100 # to get feasable numbers for dimensions
lp = Dis*X
lp = round(l*100.0)/100.0

print 'Stator Inner Diameter = ', Dis*1000.0,'mm'
print 'Equivalent Axial Length = ', lp*1000.0,'mm'
```

Here, there is a mechanical constraint that we should consider. It is tip speed of the rotor. We need to calculate the tip speed of the rotor for the maximum rotor speed and confirm that it is sufficiently small. Rated speed of the train is 72 km/h and the maximum speed is given as 140 km/h. Then, the maximum rotor speed in rpm is given by:

```python
nmax=nrated*140.0/72.0 #rpm
```

Tip speed of the rotor for the maximum speed approxiamtely equals to:

```python
tipspeed = (nmax/60.0)*pi*Dis #m/s
print 'Tip speed = ', tipspeed,'m/s'
```

The resulting tip speed is less than 100 m/s and it is in the acceptable range.

Determining the inner stator diameter and the axial length of the motor, the next step will be assigning stator outer diameter and airgap values.

Stator outer diameter for a 6 pole machine can be found by using this ratio:

```python
from IPython.display import Image 
Image(filename='DosDis.PNG')
```

```python
DosoverDis = 1.78
Dos = Dis*DosoverDis
Dos = round(Dos*100.0)/100.0 # to get feasable numbers for dimensions
DosoverDis = Dos/Dis

print 'Stator Outer Diameter = ', Dos*1000.0,'mm'
print 'Stator Inner/Outer Diameter Ratio = ', DosoverDis
```

** Determination of the airgap **

The airgap value for a 6 pole machine is given by the following equation. Note that airgap value should be at least 0.2 because of mechanical constraints.

g = 0.18 + 0.006(Pmech)^0.4

```python
g = 0.18 + 0.006*((Pmech)**0.4) #in mm
g = round(g)
g = g/1000.0 # in m
print 'Airgap = ', g*1000.0,'mm'
```

** Winding selection **

Determining the outer stator diameter and the airgap of the motor, the next step is to assign number of stator slots and form a winding diagram.

Number of slots (Q) must be a multiple of number of phases and number of poles, which is 18 in our case. Slot pitch should be 7 mm at least due to manifacturing reasons.

```python
from IPython.display import Image 
Image(filename='slotpitch.PNG')
```

```python
innerstatorcircumferance = Dis*pi
innerstatorcircumferance/72
```

The slot pitch calculated is well in the range given in the above figure. Therefore number of stator slots can be assigned as 72.

```python
Qs = 72.0
qs = Qs/(polepair*2)/3.0 #number os slots per pole per phase
```

We can use double layer windings with 8/9 winding pitch to eliminate 9th harmonics naturally. Then winding factor for the stator becomes:

```python
# kws = kps * kds  short pitch and distribution factors.
kps = sin(8.0*pi/18.0)
kds = sin(pi/6)/(qs*sin(pi/(6*qs)))
kws = kps * kds
print 'Stator winding factor = ', kws
```
