#  Design of a Train Traction Motor


The motor you need to design is a traction asynchronous squirrel cage induction motor with the following specifications:

Rated Power Output: 1280 kW

Line-to-line voltage: 1350 V

Number of poles: 6

Rated Speed: 1520 rpm (72 km/h) (driven with 78 Hz inverter)

Rated Motor Torque: 7843 Nm

Cooling: Forced Air Cooling

Insulating Class: 200

Train Wheel Diameter: 1210 mm

Maximum Speed: 140 km/h

Gear Ratio: 4.82

### 1) ELECTRICAL DESIGN PARAMATERS

```python
Pout=1280e3
Vline=1350.0 # Line to line Voltage
Vln=Vline/sqrt(3) # Line to neutral Voltage
p1=3.0 # Number of pole pair
n_r=1520.0 # rated motor speed
f=78.0 #frequency of motor suplly
n_s=120*f/(2*p1) # synchronus speed
slip=(n_s-n_r)/n_s
Trated=7843.0 # Output Torque

```

### 2)DIMENSIONS OF STATOR CORE

```python
Pole_Power=Pout/(2*p1)
print 'Pole_Power=',Pole_Power,'W'
Cmech=270.0 # From lecture notes for 6-pole machine
print 'Cmech=',Cmech,'kW S/m^3'
#x=l/Din
x=(p1**0.3333*pi)/(2*p1)
print 'l/Din=',x
#Cmech=Pout/Din^2*l*nsynch=Pout/Din^3*x*nysnch
nsynch=f/p1 #synchronous rotor speed in Hertz
print'nsynch=',nsynch,'Hz'
Din_stator=(Pout/1000/(Cmech*nsynch*x))**0.3333*1000
print 'Din_stator=',Din_stator,'mm'
l=Din_stator*x
print'l=', l,'mm'

```

```python
from IPython.display import Image 
Image(filename='Din_Dout.png')
```

```python

Dout_stator=Din_stator/0.7
print 'Dout_stator=',Dout_stator,'mm'
g=0.18+0.006*Pout**0.4 #Air Gap calculation from lecture notes
print '(Air Gap)=g=',g,'mm'
T_PolePitchStator=pi*Din_stator/(2*p1)
print 'T_PolePitchStator=',T_PolePitchStator,'mm'
# Lets choose q=4 (Number of slots per phase per pole)
q=4
T_SlotPitchStator=T_PolePitchStator/(3*q)
print 'T_SlotPitchStator=',T_SlotPitchStator,'mm'
Qs=q*3*2*p1 # Number of slots of Stator
print '(Number_of_slots_stator)=Qs=',Qs,
```

### 3)STATOR WINDING DESIGN

#### A)Winding Factor

```python
# Lets use 10/12 pitch double layer winding to eliminate 5th harmonics 
#Note that it does not eliminate totally but the effect of 5th harmonic is decreased
#kp:pitch factor or chording factor#
#kd:distribution factor or zone factor

```

Winding factor for fundamental harmonic is calculated below

```python
kp=sin((pi/2)*10/12)
print 'kp=',kp
#kd=sin(q*alpha/2)/q*sin(alpha/2)
alpha=pi/(Qs/(2*p1))
kd=sin(q*alpha/2)/(q*sin(alpha/2))
print 'kd=',kd
kw=kd*kp
print 'kw=',kw
```

#### B)Number of turns per phase and Number of conductor per slot

For magnetic loading, below table is used as reference.

```python
from IPython.display import Image 
Image(filename='Flux_density_Rates.png')
```

Flux density of air gap should be between 0.7 T and 0.9 T for 6 pole machine in order to satisfy saturation constraints.

```python
Bgap_max=0.75
Bgap_average=Bgap_max*(2/pi)
#Pole_Flux=Bgap*Apole
Apole_stator=T_PolePitchStator/1000*l/1000


Pole_Flux=Bgap_average*Apole_stator
print 'Pole_Flux=',Pole_Flux,'Wb'
```

```python
#where E_stator=Vline/sqrt(3)=Vln and W1 :Number of turn per phase
Em=0.98*Vln #induced voltage at stator

Ns=(Em*sqrt(2))/(2*pi*f*Pole_Flux)
print 'Number of turns per phase_analytical)=Ns=',Ns


```

```python
Ns=36.0
a=2 # number of parallel path
zQs=(2*a*3*Ns)/Qs #Number of conductor per slot  it should be even and integer
print '(Number of conductors per slot_analytical)=zQs=',zQs
```

Therefore;

```python
Ns=(Qs*zQs)/(2*a*3)
print'(Number of conductors per slot_real)=zQs=', zQs
print'(Number of turns per phase_analytical)=Ns=',Ns
```

```python
Bgap=(30.107/36.0)*Bgap_max
print 'Bgap=',Bgap
```

#### D)Conductor Size and Selection of Stator Winding

```python
from IPython.display import Image 
Image(filename='motor_efficiency.png')
```

```python
#Accordibg to above efficiency table  I choose 1280 kW motor efficiency=95%
Eff=0.96
pf=0.9 # Assumed power factor
Iln_s=Pout/(Eff*pf*sqrt(3)*Vline) #stator line to neutral current
print 'Iln_s=',Iln_s,'A'
```

```python
J=7 #A/mm2 for 6 pole machine
Aco_s=Iln_s/(J*a)
print'Aco_s=',Aco_s,'mm^2'
Dco_s=sqrt(Aco_s*4/pi) # Diameter of stator side conductor
print'Dco_s=', Dco_s,'mm'
```

```python
#Skin Depth Calculation for Copper cable at 78 Hz
#p=1.68*10**-8 # Copper resistivity at 20 'C
#c=0.00386 # Temperature coefficient of copper per degree
p=1.68*10**-8 
w=2*pi*f
u=4*pi*0.999991e-7
Skin_Depth=sqrt((2*p) / (w*u))*1000 # in mm
print 'Skin_Depth=',Skin_Depth,'mm'
```

### 5)STATOR SLOT DIMENSIONS

Firstly, below stator geometry is selected

Then,these slot paramaters will be calculated

```python
from IPython.display import Image 
Image(filename='Stator_slot.png')
```

```python
Kfill=0.44 # for above 10 kw Motor
Asu=(pi*(Dco_s**2)*ap*zQs)/(4*Kfill)
print 'Stator Slot Area=Asu=',Asu,'mm^2'
```

```python
bos=2.5 # mm # from lecture handout < 8*air gap distance
hos=0.75 # mm # from lecture handout
hw=2 # mm # from lecture handout
Bts=1.6 #T# aimed tooth flux density 1.5 T<Bts<1.65 T
Kfe=0.96 # Lamination Factor for 0.5 mm thick lamination
bts=(Bgap*T_SlotPitchStator)/(Bts*Kfe)
print 'bts=',bts,'mm'
bs1=((pi*(Din_stator+2*hos+2*hw))/ Qs)-bts 
print 'bs1=',bs1,'mm'
```

In order to find bs2 and hs following equations are used

1)Asu=hs*(bs1+bs2)/2

    2*Asu/hs=(bs1+bs2)

2)bs2=bs1+2*hs*tan(pi/Ns)

    bs2-bs1=2*hs*tan(pi/Ns)
    
    When we multipy these equations, we obtain
    (bs2^2-bs1^2)=4*Asu*tan(pi/Ns)
    bs2=sqrt(4*Asu*tan(pi/Ns)+bs1^2)

```python
bs2=sqrt(4*Asu*tan(pi/Qs)+bs1**2)
print'bs2=' ,bs2,'mm'
hs=(2*Asu)/(bs1+bs2)
print 'hs=',hs,'mm'
hcs=(Dout_stator/2)-(Din_stator/2)-(hos+hw+hs)
print 'hcs=',hcs,'mm'
```

```python
Bcs_max=((pi/2)*Pole_Flux/(2*l*hcs))*1e6 # Back Core Flux density

print 'Bcs_max=',Bcs_max,'T'
```

**Attention!!!** :Back Core flux is very low, for efficient design Bcs should be between 1.4 T and 1.7 

Therefore, Dout_stator should be decreased until these values are obtained.

Dout_stator=889 mm which is calculated in part-2(Decrease it!!) and calculate hcs again!!

```python
Dout_stator_updated=800 #mm
hcs_updated=(Dout_stator_updated/2)-(Din_stator/2)-(hos+hw+hs)
print 'hcs=',hcs_updated,'mm'
Bcs_updated=(Pole_Flux/(2*l*hcs_updated))*1e6
print 'Bcs=',Bcs_updated,'T'

```

### 7) THE MAGNETIZING CURRENT

Firstly, Carter's coefficients of stator and rotor is calculated

```python
Ys=(bos**2)/(5*g+bos)
print 'Ys=', Ys,'mm'
Yr=(bor**2)/(5*g+bor)
print 'Yr=', Yr,'mm'
```

```python
Kcs=T_SlotPitchStator/(T_SlotPitchStator-Ys) # carter's coefficient of stator
print 'Kcs=',Kcs
Kcr=T_slotPitchRotor/(T_slotPitchRotor-Yr) # carter's coefficient of rotor
print 'Kcr=',Kcr
Kc=Kcs*Kcr # total carter^s coefficient
print 'Kc=',Kc
```

```python
from IPython.display import Image 
Image(filename='magnetization_curve.png')
```

```python
Fmg=((Kc*g*Bgap)/(4*pi*1e-7))/1000 #airgap MMF
print 'Fmg=',Fmg,'Aturns'
#Bts=1.6 T and H=2460 A/m
Fmts=2460*(hs+hw+hos)/1000 #stator tooth MMF
print'Fmts=', Fmts,'Aturns'
#Btr=1.6 T and H=2460 A/m
Fmtr=2460*(hr+hor)/1000 #rotor tooth MMF
print'Fmtr=', Fmtr,'Aturns'
```

```python
#Ccs:Subunitary  Emprical Coefficient of Stator
#Ccr:Subunitary  Emprical Coefficient of Rotor
#Bcs=1.63 T
#Bcr=1.6 T
Ccs=0.88*exp(-0.4*(Bcs_updated**2)) 
Ccr=0.88*exp(-0.4*(Bcr**2)) 
Hcs=3060 # A/m #From magnetization curve
Fmcs=Ccs*(pi*(Dout_stator_updated-hcs_updated)*Hcs)/(2*p1)/1000 #Stator back core MMF
Hbr=2460 #From magnetization curve#
Fmcr=Ccr*(pi*(D_shaft_max+hcr)*Hbr/(2*p1))/1000 #Rotor back core MMF
print 'Fmcs =',Fmcs,'Aturns'
print 'Fmcr =',Fmcr,'Aturns'
```

```python
F1m=2*(Fmg+Fmts+Fmtr+Fmcs+Fmcr) #Magnetizing MMF
print 'F1m=',F1m,'Aturns'
```

```python
Im=(pi*p1*F1m)/(2*3*sqrt(2)*Ns*kw) #magnetizing current
print 'Im=',Im,'A'
Im_pu=Im/Iln_s
print 'Im_pu=',Im_pu,'pu'
```

### 8) RESISTANCE AND INDUCTANCES

**A) Stator Resistance**

```python
y=5*T_SlotPitchStator/6 #coil span
print 'y=',y,'mm'
l_end=(pi/2)*y+18 #end connection length
print 'l_end=', l_end,'mm'
l_coil=2*(l+l_end) # coil length
print 'l_coil=',l_coil,'mm'
```

```python
p_co_20=1.78*1e-8 #copper resistivity at 20 C
p_co_80=p_co_20*60/273+p_co_20 #copper resistivity at 80 C #Assumed stator temperature
Rs=(p_co_80*l_coil*Ns)*1000/(Aco_updated*a) #Stator resistance
print 'Rs=',Rs,'Ohm'

```

**B) Rotor Resistance**

```python
p_Al_20=3.1*1e-8 #Aluminium resistivty at 20 C # rotor bar is aluminium
p_Al_80=p_Al_20+p_Al_20*60/273 # Aluminium resistivty at 80 C 
print p_Al_80
KR=1.74
#Assume that  hr*alpha_rotor is equal to Kr based on handout equation 15.73
print 'KR=',KR #skin effect resistance coefficient
```

```python
l_er=(pi*(Der-b_end_ring))/Qr # end ring segment length
print 'l_er=',l_er,'mm'
Rbe=(p_Al_80*l*KR/Aru+(p_Al_80*l_er)/(2*Aring*sin(pi*p1/28)*sin(pi*p1/28)))*1000 # bar/end ring segment equivalent resistance
print 'Rbe=',Rbe,'Ohm'
```

```python
Rr=4*3*((Ns*kw)**2)*Rbe/Qr #  stator referred rotor resistance 
print 'Rr=',Rr,'Ohm'
```

**C) Stator Leakage Inductance**


```python

Lamda_s=(((hs/(bs1+bs2))+(2.0*hw/(bos+bs1))+(hos/bos)))*(2/3.0)*((1+3*(0.833))/4.0)#Slot leakage coefficient 
print 'Lamda_s =',Lamda_s
Cs=1-(0.0033*((bos**2)/(g*T_SlotPitchStator)))
print 'Cs=',Cs
phi=pi*(6*5/6-5.5)
print 'phi=',phi,'rad'
Gama_ds=(0.14*sin(phi)+0.76)*1e-2 # for q=4 from handout 15.78
print 'Gama_ds=',Gama_ds
Kst=(Fmts+Fmtr)/Fmg #saturation factor
print 'Kst=',Kst
print 'Kc=',Kc #Carter Coefficient
Lamda_ds=(0.9*T_SlotPitchStator*q*q*kw*kw*Cs*Gama_ds)/(Kc*g*(1+Kst))
print 'Lamda_ds=',Lamda_ds #  stator differential permeance coefficient
Lamda_ec=(0.34*q*(l_end-0.64*0.833*T_SlotPitchStator))/l
print 'Lamda_ec=', Lamda_ec # rotor end connection specific geometric permeance
L_sl=(2*(4*pi*1e-7)*l*Ns*Ns*(Lamda_s+Lamda_ds+Lamda_ec))/(1000*p1*q)
print 'L_sl=',L_sl,'H' # Stator leakage inductance
X_ls=2*pi*f*L_sl
print'X_ls=',X_ls,'Ohm'
```


**D) Rotor Leakage Inductance**

```python
Lamda_r=0.66+(2*hr)/(b1r+b2r)+hor/bor #from handout equation 15.81
print'Lamda_r=',Lamda_r #rotor dlot leakage coefficient
Gama_dr=9*((p1*6/Qr)**2)*1e-2
print 'Gama_dr=', Gama_dr
Lamda_dr=(0.9*T_slotPitchRotor*Gama_dr*Qr*Qr)/(6*6*p1*p1*Kc*g)
print 'Lamda_dr=',Lamda_dr #rotor differential permeance coefficient
Lamda_er=(2.3*(Der-b_end_ring)*log10((4.7*(Der-b_end_ring))/(b_end_ring+2*a_end_ring)))/(Qr*l*4*sin(pi*p1/Qr)*sin(pi*p1/Qr))
print 'Lamda_er=',Lamda_er # rotor end ring permeance coefficient
Kx=3/(2*KR) #from handout skin effect coefficient fot leakage reactance
print'Kx=', Kx
X_be=2*pi*f*(4*pi*1e-7)*l*((Lamda_r*Kx)+Lamda_dr+Lamda_er)/1000 #bar leakage reactance
print 'X_be=',X_be,'Ohm'
X_rl=4*3*Ns*Ns*kw*kw*X_be/Qr # Stator referred rotor leakage reactance
print 'X_rl=',X_rl,'Ohm'


```

**E)Magnetizing Inductance**

```python
Xm=(sqrt((Vln/Im)**2-Rs**2))-X_ls
print 'Xm=',Xm,'Ohm'
```

**F)Skew Effect**

To decrease harmonics on rotor side,rotor slots are one skewed.


```python
K_skew=(sin(pi/(2*3*q)))/(pi/(2*3*q))  #rotor skew coefficient from handout equation 15.90
print 'K_skew=',K_skew
```

```python
Xm_skew=Xm*K_skew
print 'Xm_skew=',Xm_skew,'Ohm'
```

```python
X_rl_skew=Xm*(1-(K_skew**2))+X_rl
print 'X_rl_skew=',X_rl_skew,'Ohm'
```

### 9)LOSSES AND EFFICIENCY

**A)Copper Losses**

copper is used on stator winding coils

```python
Pcu_s=3*(Iln_s**2)*Rs #Copper loss due to stator winding resistance
print 'Pcu_s=',Pcu_s,'W'
```

**B)Aluminium Loss**

Aluminium is used on rotor bar and rotor end ring



```python
Pal_r=3*(Iln_s**2)*Rr #Copper loss due to stator winding resistance
print 'Pal_r=',Pal_r,'W'
```

**C) Mechanical /Ventilation Losses**

```python
Pmv=0.008*Pout # from handout for 6 pole IM.
print 'Pmv=',Pmv,'W'
```

**D)Stray Loss**

```python
Pstray=0.01*Pout # from handout for 6 pole IM.
print 'Pstray=',Pstray,'W'
```

**E)Iron Losses**

**E-1) Fundamental Iron Losses**

```python
p10=2 # W/kg from catalog at 50 Hz at 1 Tesla
d_iron=7800 # kg/m^3
```

```python
G_stator_teeth=(d_iron*Qs*bts*(hs+hw+hos)*l*Kfe)/1e9
print 'G_stator_teeth=',G_stator_teeth,'kg'
```

```python
Kt=1.6 # mechanical machining loss coefficient for stator tooth
Piron_stator_teeth=Kt*p10*((f/50)**1.3)*(Bts**1.7)*G_stator_teeth
print 'Piron_stator_teeth=',Piron_stator_teeth,'W'
```

```python
G_stator_yoke=(d_iron*pi*(Dout_stator_updated**2-(Dout_stator_updated-hcs_updated)**2)*l*Kfe)/(1e9*4)
print 'G_stator_yoke=',G_stator_yoke,'kg'
```

```python
Ky=1.6 # mechanical machining loss coefficient for stator back core(yoke)
Piron_stator_yoke=Ky*p10*((f/50)**1.3)*(Bcs_updated**1.7)*G_stator_yoke
print 'Piron_stator_yoke=',Piron_stator_yoke,'W'
```

**E-2)Harmanoics Losses**

```python
Kps=1/(2.2-Bts)
Kpr=1/(2.2-Btr)
print Kps
Bps=(Kcr-1)*Bgap
print 'Bps=',Bps,'T'
Bpr=(Kcs-1)*Bgap
print 'Bpr=',Bpr,'T'
```

```python
G_rotor_teeth=(d_iron*l*Kfe*Qr*hr*btr)/1e9
print'G_rotor_teeth=',G_rotor_teeth,'kg'
```

```python
Piron_harmonics=(0.5*1e-4*((Qr*f*Kps*Bps)/p1)**2*G_stator_teeth+0.5*1e-4*((Qs*f*Kpr*Bpr)/p1)**2*G_rotor_teeth)
print 'Piron_harmonics=',Piron_harmonics,'W'
```

```python
Piron=Piron_stator_teeth+Piron_stator_yoke+Piron_harmonics
print 'Piron=',Piron,'W'
```

**F)Efficiency**


```python
Efficiency=(Pout/(Pout+Pcu_s+Pal_r+Pmv+Pstray+Piron))*100
print'Efficiency=',Efficiency,'%'
```
