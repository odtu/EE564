#Main Dimensions of Stator Core

\begin{equation}
  D_{is} = \sqrt[3]{\frac{2p*p_1*S_{gap}}{pi*\lambda*f_1*C_0}}
\end{equation}

```python
P_n = 150 
V_1 = 1000
n_1 = 1800 
p   = 2
p_1 = 2
f_1 = 50
cos_theta = 0.85
K_E = 0.98 - 0.005*p_1
eff = 0.9
q=3
```

\begin{equation}
  S_{gap} = \frac{K_E*P_n}{\eta*\cos(\theta_n)}
\end{equation}

```python
print K_E
```

```python
S_gap = ((K_E*150*1000)/(eff*cos_theta))
print S_gap
```

###for a 4 pole machine,
\begin{equation}
\lambda = 1.1
\end{equation}

```python
import math
D_is = math.pow((2*p*p_1*S_gap)/(3.14*1.1*f_1*265000),1.0/3.0)
print D_is
```

\begin{equation}
D_{is} = 0.32 m
\end{equation}

\begin{equation}
\tau = \frac{pi*D_{is}}{2*p}
\\\tau_{s} = \frac{\tau}{3*q}
\end{equation}

```python
pole_pitch = 3.14*D_is/(2*p)
print pole_pitch
```

\begin{equation}
\tau = 0.25 m
\end{equation}

```python
slot_per_pole= pole_pitch/(3*q)
print slot_per_pole
```

####again for a 4 pole machine;
\begin{equation}
\frac{D_{is}}{D_{out}} = K_D = 0.62
\end{equation}

```python
D_out = D_is/0.62
print D_out
```

```python
D_out=0.53
```


\begin{equation}
D_{out} = 0.52 m
\end{equation}

\begin{equation}
g = (0.1 + 0.012*\sqrt[3]{P_n}*10^{-3})
\\for \quad 2p>2
\end{equation}

```python
import math
g = (0.1+0.012*math.pow(P_n*1000,1.0/3.0))/(1000)
print g
```

\begin{equation}
g = 0.7 mm
\end{equation}

#The Stator Winding 

\begin{equation}
N_s = 2*p*q*m = 2*2*3*3 = 36
\end{equation}

\begin{equation}
\alpha_{ec} = \frac{2\pi*p}{N_s} = \frac{\pi}{9}
\end{equation}

\begin{equation}
K_{ql} = \frac{\sin(\frac{\pi}{6})}{q*\sin(\frac{\pi}{6q})}
\end{equation}

```python
import math
K_ql = math.sin(3.14/6)/(3*math.sin(3.14/18))
print K_ql
```

\begin{equation}
for\quad chording : \frac{y}{\tau} = \frac{7}{9}\quad is \quad selected
\end{equation}

\begin{equation}
so\quad the\quad electrical\quad angle\quad is:\quad \frac{2*\pi*p}{N_s}=\quad \frac{\pi}{9}
\end{equation}

\begin{equation}
chording\quad factor\quad is:\quad \sin(\frac{pi*y}{2*\tau})
\end{equation}

```python
import math
K_yl = math.sin((3.14*7)/(2*9))
print K_yl
```

```python
import math
Kw1= K_yl*K_ql
print Kw1
```

####following values are assumed from now on:


```python
B_g = 0.7 
alpha_i = 0.65
L=1.1*3.14*D_is/(2*p)
rated_current = 265000/(1.1*alpha_i*Kw1*3.14*B_g*p)
print rated_current/1000
```

```python
pole_flux=alpha_i*pole_pitch*L*B_g
print pole_flux
```

####number of turns per phase W1:
\begin{equation}
\frac{K_E*V_1}{4*K_f*K_{W1}*f*\Phi}
\end{equation}

```python
import math
W_1 = (K_E*V_1)/(4*1.1*Kw1*50*pole_flux*math.sqrt(3))
print W_1 
```

####number of conductors ns:
\begin{equation}
ns:\quad \frac{W_1}{p*q}
\end{equation}
#### (current paths in parallel:1)

```python
ns = W_1/(p*q)
print ns
```

```python
ns = 14
```

```python
import math
rated_current= 150000/(eff*cos_theta*math.sqrt(3)*V_1)
print rated_current
```

```python
J_co=6.3
```

####magnetic wire cross section:
\begin{equation}
A_{co}=\quad \frac{I_{rated}}{J_{cos}}
\end{equation}

```python
A_co=rated_current/J_co
print A_co
```

\begin{equation}
A_{co}=conductor\quad width*conductor\quad height
\end{equation}
####a rectengular cross section conductor will be used,so standart value of 10 mm is taken here for slot width

```python
slot_width = 10
conductor_insulation = 0.4
slot_insulation = 4
conductor_width = slot_width - conductor_insulation - slot_insulation
print conductor_width
```

#### conductor height = A_co/conductor width = 18/5.6 = 3.21mm

```python
radial_cond_insulation = conductor_insulation*ns
radial_slot_insulation = slot_insulation*2
interlayer_insulation = 2
wedge = 4
radial_insulation = radial_cond_insulation + radial_slot_insulation+interlayer_insulation+wedge
print radial_insulation
```

####slot height:
\begin{equation}
h_s = h_{ins} + n_s*b_c
\end{equation}

```python
bc=3.21
h_s = radial_insulation + ns*3.21
print h_s
```

####back iron radial thickness:
\begin{equation}
h_{cs} = \frac{D_{out}-D_{is}}{2}-h_s
\end{equation}

```python
h_cs= ((D_out-D_is)*1000)/2-h_s
print h_cs
```

####so all in all;
\begin{equation}
h_s = 64.54 mm
\\h_{cs} = 39.7 mm
\end{equation}
####thus back iron flux density is:
\begin{equation}
B_{cs} = \frac{\Phi}{2*l_i*h_cs}
\end{equation}
####where l_i is stack length, and calculated as 0.277mm

```python
B_cs = pole_flux/(2*L*0.001*h_cs)
print B_cs
```

#Rotor performance calculation

####magnetization mmfs:

\begin{equation}
F_{lm}=F_g+F_s+(F_{st}+F_{rt})+\frac{F_{cs}}{2}+\frac{F_{cr}}{2}
\end{equation}

\begin{equation}
F_g=K_c*g*\frac{B_g}{\mu_o}
\\K_c=K_{c1}*K_{c2}
\\K_{c1,c2}=\frac{\tau_{s,r}}{\tau_{s,r}-\gamma_{s,r}g}
\\gamma_{s,r}=\frac{\frac{b_{os,r}}{g}^2}{5+\frac{b_{os,r}}{g}}
\end{equation}

```python
import math
gamma=math.pow(slot_width,2)/(5*g*1000+slot_width)
print gamma
```

```python
Kc2= rotor_tooth_pitch/(rotor_tooth_pitch - gamma*g)
print Kc2
```

```python
Kc1=1.3
```

```python
Kc=Kc1*Kc2
print Kc
```

```python
import math
Fg = (Kc*g*B_g)/(1.256*math.pow(10,-6))
print Fg
```

```python
h_st=40
Ns=36
import math
Btsmin=math.pow(3.14*(D_is*1000+2*h_st*2)-Ns*slot_width,-1)*B_g*3.14*D_is*1000
print Btsmin
```

```python
h_rt=25
import math
Btrmax=3.14*(D_is*1000-2*g*1000)*B_g/(3.14*(D_is*1000-2*g*1000-2*h_rt)-Nr*br)
print Btrmax
```

```python
Btsmed=(3.14*D_is*1000*B_g)/((3.14*(D_is*1000+h_st)-Ns*slot_width))
print Btsmed
```

```python

Btrmed=(3.14*D_is*1000*B_g)/(3.14*(D_is*1000-h_rt)-Nr*br)
print Btrmed
```

####from lamihation magnetization curve;
Hstmin=106 A/m, Hstmed = 177 A/m, Hstmax= 585 A/m
####and also;
\begin{equation}
F_{st}=h_{st}*H_{st}
\end{equation}

```python
Hts=(106+4*177+585)/6
print Hts
```

```python
Fst=h_st*0.001*Hts
print Fst
```

```python
Btrmin = 1.15
Htrmin=310
Htrmed=356
Htrmax=760
Htr=(Htrmin+4*Htrmed+Htrmax)/6
print Htr
```

```python
Frt=Htr*h_rt*0.001
print Frt
```

####Stator back core mmf Fcs is:
\begin{equation}
F_{cs}=l_{cs}*H_{cs}
\end{equation}

```python
Bcsmax = (B_g*rotor_tooth_pitch*0.01)/(3.14*(D_out-D_is)/2)
print Bcsmax
```

```python
Hcsmax= 177
```

####the mean length lcs is:
\begin{equation}
l_{cs} = \frac{\pi*(D_{out}-h_{cs})}{2p}
\end{equation}

```python
lcs=3.14*(D_out-h_cs*0.001)/(2*p)
print lcs
```

```python
Fcs=lcs*Hcsmax
print Fcs
```

####same equations followed by rotor back iron mmf:

```python
Hcr=237
D_ir=D_is-2*g*0.001-2*h_rt*0.001-2*h_cs*0.001
print D_ir
lcr=3.14*(D_ir-h_cr*0.001)*1.4/(2*p)
print lcr
Fcr=Hcr*lcr
print Fcr
```

####so the total magnetization mmf becomes:


```python
Flm=Fg+(Frt+Fst)+Fcr/2+Fcs/2
print Flm
```

\begin{equation}
I_{\mu}=\frac{F_{lm}*p*\pi}{3\sqrt(2)*W_1*K_{w1}}
\end{equation}

```python
import math
magnetization_current = Flm*p*3.14/(3*math.sqrt(2)*W_1*Kw1)
print magnetization_current
```

```python
current_ratio=magnetization_current/rated_current
print current_ratio
```

#Rotor winding parameters

```python
import math
k01= 1 - 0.033*slot_width*slot_width/(g*1000*slot_per_pole*1000)
kapa01=0.003
differential_permeance=0.9*slot_per_pole*1000*q*Kw1*q*Kw1*k01*kapa01/(Kc*g*1000)
print differential_permeance
```

```python
l_i=0.277
l1=0.015
import math
lfs=2*(l1+(7.0/9.0)*pole_pitch/math.sin(0.52))+3.14*h_s*0.001
print lfs
end_permeance=0.34*(q/l_i)*(lfs-0.64*(7.0/9.0)*pole_pitch)
print end_permeance
```

```python
X_sl=15.8*(f_1/100.0)*(W_1/100)*(W_1/100)*(l_i/(p*q))*(slot_leakage+differential_permeance+end_permeance)
print X_sl
```

```python
import math
K_80_deg=2.1712
Lgeo=l_i+6*0.0075
R_s=(W_1/A_co)*(Lgeo+lfs)*K_80_deg*math.pow(10,-2)
print R_s
```

```python
R_r=0.8*R_s
print R_r
```

```python
Rr_r=(K_80_deg*2*(lfr+Lgeo)*W_2/A_cor)*0.01
print Rr_r
```
