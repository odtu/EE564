#Main Dimensions of Stator Core

\begin{equation}
  D_{is} = \sqrt[3]{\frac{2p*p_1*S_{gap}}{pi*\lambda*f_1*C_0}}
\end{equation}

```python
P_n = 150 
V_1 = 1000
n_1 = 1800 
p   = 2
p_1 = 2
f_1 = 50
cos_theta = 0.85
K_E = 0.98 - 0.005*p_1
eff = 0.9
q=3
```

\begin{equation}
  S_{gap} = \frac{K_E*P_n}{\eta*\cos(\theta_n)}
\end{equation}

```python
print K_E
```

```python
S_gap = ((K_E*150*1000)/(eff*cos_theta))
print S_gap
```

###for a 4 pole machine,
\begin{equation}
\lambda = 1.5
\end{equation}

```python
import math
D_is = math.pow((2*p*p_1*S_gap)/(3.14*1.5*f_1*147000),1.0/3.0)
print D_is
```

\begin{equation}
D_{is} = 0.35 m
\end{equation}

\begin{equation}
\tau = \frac{pi*D_{is}}{2*p}
\\\tau_{s} = \frac{\tau}{3*q}
\end{equation}

```python
pole_pitch = 3.14*D_is/(2*p)
print pole_pitch
```

\begin{equation}
\tau = 0.27 m
\end{equation}

```python
sl_per_pole= pole_pitch/(3*q)
print sl_per_pole
```
