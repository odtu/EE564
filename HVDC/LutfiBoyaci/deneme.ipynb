```python
S=6500000.00 #VA
V1=3000.00 #V
f=500
V2=300000.00  #V
I2= S/V2 #A
print 'I2 =', I2 ,'A'
```

```python
Turn_ratio= 300/3
print 'Turn_ratio =', Turn_ratio
```

assume: 

```python
N1=8
N2=805 #putting a guarantee margin to secondary turn ratio
n=0.98 #efficiency # to compansate magnetizing losses and other losses
```

```python
I1=S/(n*V1)
print 'I1 =', I1 ,'A'
```

assume:

```python
ccc=4 #A/mm^2       #Current carrying capacity of wire
A2= I2/ccc          # needed area of secondary conductor
print 'needed area of secondary conductor=', A2 ,'mm^2'
r2 = sqrt(A2/pi)    # radius of secondary conductor
print 'radius of secondary conductor =', r2 ,'mm'
A1= I1/ccc          # needed area of single primary conductor
print 'needed area of single primary conductor=', A1 ,'mm^2'
r1 = sqrt(A1/pi)    # radius of secondary conductor
print 'radius of secondary conductor =', r1 ,'mm'
```

###### Skin Depth:

```python
p=1.68e-08*(1+0.0039*(110-20)) #resistivity for copper @ 110*C
w=2*pi*500
u=4e-07*pi*0.999991 # permeability of air*relative mag. permeability of copper
SD=sqrt(2*p/(w*u))*1000 #skin depth in mm
print 'Skin Depth =', SD ,'mm'
```

Considering the skin depth, maximum radius of the conductor should be smaller than the 3.39mm. Needed radius of secondary conductor is 1.313mm and it is appropriate for our consideration. From the AWG Cables catalogue, we can choose the AWG9 as a secondary side conductor.

```python
# characteristics of AWG9 cable:
Awg9_diameter=2.906 #mm
Awg9_resistance= 2.599 #mohm/m
```

However, there is no available conductor size to carry all the primary side current because of high current. Therefore, more than one conductor will be used as parallel to carry primary side current. Due to skin effect, cable having a diameter smaller than 6.78 mm (2xskin depth) should be chosen as a primary conductor.  From the catalogue, AWG2 cable is chosen as a primary side conductor.

```python
# characteristics of AWG2 cable:
Awg2_diameter=6.544 #mm
Awg2_resistance= 0.5127 #mohm/m
Awg2_area= 33.6 #mm2
Iawg2=ccc*Awg2_area
primary_no_of_conductor=round(I1/Iawg2 , 0)
print (primary_no_of_conductor)
```

###### Finding Core Area:

```python
B= 1.7 #Tesla
kw= 0.945 #lamination factor
#V2=4.44*f*N2*kw*B*Acore
Acore= V2/(4.44*f*N2*kw*B)
print 'Core Area=' , Acore , 'm^2' 
core_edge=sqrt(Acore)  #assuming square core area
print 'Core Edge=', core_edge , 'm'
```

###### Finding Core Loss:

```python
#23JGSD080 is chosen as a core material from the catalogue
#datasheet parameters:
#Core loss @ 1.7T and 50Hz= 0.75 W/kg
#Core loss @ 1.7T and 60Hz= 0.97 W/kg

#Steinmetz's equation: Q=k*(f^a)*(B^b)
Core_loss_50= 0.75
Core_loss_60= 0.97
a=log(Core_loss_60/Core_loss_50)/log(60.0/50.0)
Core_loss_500=Core_loss_50*(500.0/50.0)**a
print 'Core Loss @ 500Hz=', Core_loss_500 , 'W/kg'
```

###### Core Window Area:

```python
kw1= 0.3 # primary winding factor
kw2= 0.3  ## secondary winding factor
Area_awg9= 6.63 #secondary cable area mm^2
Area_awg2= 33.6 #primary cable area in mm^2
primary_window_area = primary_no_of_conductor*Area_awg2*N1/kw1/100 #in cm^2
secondary_window_area = Area_awg9*N2/kw2/100 #in cm^2
print 'primary window area=' , primary_window_area , 'cm^2'
print 'secondary window area=' , secondary_window_area , 'cm^2'
total_window_area=primary_window_area+secondary_window_area
print 'total window area=' , total_window_area , 'cm^2'
```
